```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
titanic_train = pd.read_csv("train.csv")
titanic_test = pd.read_csv("test.csv")
test_ids=titanic_test['PassengerId']
```

```python
def clean(data):
    data= data.drop(['Ticket','Cabin','PassengerId'],axis=1)
    cols= ['Fare','Age']
    for col in cols:
        data[col].fillna(data[col].median(),inplace=True)
    data['Embarked'].fillna(data['Embarked'].mode()[0],inplace=True)
    return data
```

```python
data=clean(titanic_train)
test=clean(titanic_test)
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
cols=['Sex','Embarked']
for col in cols:
  data[col]= le.fit_transform(data[col])
  test[col]= le.fit_transform(test[col])
  print(le.classes_)
```

```python
test
```

```python
# Define a function to extract titles
def extract_title(name):
    title = name.split(',')[1].split('.')[0].strip()
    return title

# Apply the function to create a new column 'Title'
data['Title'] = data['Name'].apply(extract_title)
test['Title'] = test['Name'].apply(extract_title)
```

```python
data['Title'].unique()
```

```python
data.head()
```

```python
test.head()
```

```python
data.drop('Name',axis=1,inplace=True)
test.drop('Name',axis=1,inplace=True)
```

```python
y= data['Survived']
X= data.drop('Survived', axis=1)
```

```python
# selecting the dataset into categorical and numerical columns
X_cat= X.select_dtypes(include='object')
X_num= X.select_dtypes(exclude='object')
test_cat= test.select_dtypes(include='object')
test_num= test.select_dtypes(exclude='object')
```

```python
X_cat['Title'].unique()
```

```python
X_cat_dummy = pd.get_dummies(X_cat,drop_first=True)
test_cat_dummy = pd.get_dummies(test_cat,drop_first=True)
```

```python
X_new=pd.concat([X_num,X_cat_dummy],axis=1)
test_new=pd.concat([test_num,test_cat_dummy],axis=1)
```

```python
X_new['Title_Dona']=0
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X_new,y, test_size=0.2, random_state=41)

```

```python
# Step 2: Split the Data
train_base, train_meta, y_base, y_meta = train_test_split(X_train, y_train, test_size=0.5, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

model1 = RandomForestClassifier()
model2 = LogisticRegression()
```

```python

# Step 3: Train the Base Models
model1.fit(train_base, y_base)
model2.fit(train_base, y_base)
```

```python

# Step 4: Generate Predictions from Base Models
preds1 = model1.predict(train_meta)
preds2 = model2.predict(train_meta)
```

```python
# Step 5: Create a New Dataset for the Meta-Model
meta_features = np.column_stack((preds1, preds2))
```

```python
# Step 6: Train the Meta-Model
meta_model = LogisticRegression()
meta_model.fit(meta_features, y_meta)
```

```python
# Step 7: Use the Stacked Model for Predictions
test_preds1 = model1.predict(X_val)
test_preds2 = model2.predict(X_val)
test_meta_features = np.column_stack((test_preds1, test_preds2))
final_preds = meta_model.predict(test_meta_features)
```

```python
from sklearn.metrics import accuracy_score
print("tree model : ", accuracy_score(y_val,final_preds))
```

```python
missing_columns = set(X_train.columns) - set(test_new.columns)
for col in missing_columns:
    test_new[col] = 0
```

```python
test_new= test_new[X_new.columns]
```

```python
# Generate meta-features for the test data using base models
test_preds1 = model1.predict(test_new)
test_preds2 = model2.predict(test_new)
test_meta_features = np.column_stack((test_preds1, test_preds2))

# Make predictions with the meta-model
submission = meta_model.predict(test_meta_features)
```
